# Textdraw

Draw with text (a fixed width font with the used glyphs is required, [APL387](https://dyalog.github.io/APL387/) is recommended).

```apl
:Namespace textdraw

    ⍝ backgrounds
    bgs←' ·░▓█'
    
    ⍝ markers
    marker0←'*×+-' ⋄ marker1←'∘○⍟⊖' ⋄ marker3←'∆∇⍙' ⋄ marker4←'⎕⍞⍠⌹⌸⌺⌷'
    marker←marker0,marker1,marker3,marker4,(diamond←'⋄'),dot←'·'
    
    ⍝ fillings
    hlevels←' ▏▎▍▌▋▊▉█' ⋄ vlevels←' ▁▂▃▄▅▆▇█' ⋄ shades←' ░▓█' ⋄ shades1←' ·░▒▓█'
    
    ⍝ bars
    bar0←' ⎕' ⋄ bar1←' █' ⋄ bar2←' ▆' ⋄ bar3←'░▓' ⋄ bar4←'○⍟' ⋄ bar5←'·|' ⋄ bar6←'⎕⌺' ⋄ bar7←'─┼'
    bars←bar0 bar1 bar2 bar3 bar4 bar5 bar6 bar7

    ⍝ block drawing
    blocks←' ▄▀█'
    blocks2←' ','▗','▖','▄','▝','▐','▞','▟','▘','▚','▌','▙','▀','▜','▛','█'
    braille←⎕UCS(0,⍳255)+⎕UCS'⠀'
    
    ⍝ scaling of coordinates
    nice←{
        (0=⎕NC'⍺')∧1<≡⍵:∇¨⍵ ⋄ 1<≡⍵:⍺∇¨⍵
        ⍺←⌊1+3⍟≢⍵ ⋄ 1=≢⍵:⍺ ∇ 0,⍵ ⋄ 0 0≡⍵:0 0
        ∧/1 ¯1∊×⍵:⍺ ∇{
            t←(-⌽0~⍨(⌈⍺÷2)⍺⍺ 0,-⍵⌿⍨0>⍵),(⌈⍺÷2)⍺⍺ 0,⍵⌿⍨0<⍵
            (⊃,⊃∘⌽,≢)t
        }⍵
        c m←10*0 1⊤10⍟(-/h l←(⌈/,⌊/)⍵)÷⍺
        w←c×1 2 2.5 4 5 10[⎕IO⌈5⌊⌊m]
        n←(h←⌈h÷w)-l←⌊l÷w ⋄ l h×←w
        l h n
    }
    scale←{
        ⍺←50 ⋄ (1<≡⍵)∧1=≢⍺:(⍺,⍨⌊⍺×5÷16)∇ ⍵
        4=≢⍺:⎕IO⌈⌈(⍵-l)÷(h-l)÷n⊣l h t n←⍺
        1∧.<≡¨⍺ ⍵:⍺ ∇¨⍵ ⋄ 1<≡⍵:⍺ ∇¨⍵
        s←2 4 8 10 12 16 20[0 9 21 41 81 101 121⍸⍺]
        n,1+s×⊃⌽n←(⌊⍺÷1.5×s)nice ⍵
    }
    
    ⍝ rectangles
    box←{
        (0=⎕NC'⍺')∧2|⎕DR⍵:⍵∇'' ⋄ 0≠⎕NC'⍺':∇⍺⍴⍵ ⋄ 2≠≢⍴⍵:∇⍉⍪⍵
        {⊃{⍺@(⊂1 1)⌽⍉⍵}/'┌┐┘└',{⍺⍪⌽⍉⍵}/'─│─│',⊂⍕⍵}⍵  ⍝ aplcart
    }
    _rect←{
        ⍺←⊢ ⋄ 2<≡⍵:(⍺⊣⌊t÷⍨n-1)∇ n⊣t n←2↓↓⍉↑⊃⍵ 
        1<≡⍵:∇⊃∘⌽¨⍺ scale ⍵ ⋄ 3≠⎕NC'⍺':(⍺⍴⍺⍺)∇∇⍵
        ⍬≡⍴⍺⍺:⍵⍴⍺⍺ ⋄ ⍵↑⊃⍪⌿,/(⌈⍵÷⍴a)⍴⊂a←⍉∘⍪⍣(2>≢⍴⍺⍺)⊢⍺⍺
    }
    _grid_←{
        2<≡⍵:(⌊t÷⍨n-1)∇ n⊣t n←2↓↓⍉↑⊃⍵
        (0=⎕NC'⍺')∧1<≡⍵:∇⊂scale ⍵
        ⍺←2 4 ⋄ (1≥≡⍺)∧1<≡⍵:(t÷⍨n-1)∇ n⊣l h t n←↓⍉↑⍺ scale ⍵
        h w←⍺ ⋄ p←(⍺⍺,(⊃⍵⍵)/⍨w-1),(w×h-1)⍴w↑⊃⌽⍵⍵ ⋄ ⍺(p _rect)⍵
    }
    grid←'┼'_grid_'─│' ⋄ dgrid←dot _grid_'' ⋄ dotgrid←dot _grid_ dot
    white←' '_rect ⋄ light←'░'_rect ⋄ dark←'▓'_rect ⋄ black←'█'_rect
    dots←dot _rect ⋄ dots2←2 2∘((2↑dot)_rect) ⋄ hdots←2 2∘((4↑2⍴dot)_rect)
    lines←' ▏'_rect

    ⍝ layers
    _at_←{
        ⍺←' ' ⋄ i←(⊂⍵⍵-⎕IO)+⍳⍴⍺⍺
        b←,((⍺∘≢¨⍺⍺)(∧⍣(~0∊⍴⍺))(⎕IO∧.≤¨⊢)∧(⊂⎕IO+⍴⍵)∧.>¨⊢)i
        (b/,⍺⍺)@(b/,i)⊢⍵
    }
    _ontop←{⍺←⊢ ⋄ ⍺(⍺⍺ _at_(⎕IO+0,⍵(⎕IO⌈-)⍥(⊃∘⌽∘⍴)⍺⍺))⍵}
    stack←{
        ⍺←1 ¯1 ⋄ R←{⍉∘⍪⍣(2>≢⍴⍵)⊢⍵}
        r←⍴¨g←R¨R ⍵ ⋄ h w←⍺×(⌈/⊃¨r)(⌈⌿⊃∘⌽¨r)
        ⊃⍪⌿,/(h∘.,w)↑¨g
    }
    triml←{⍵/⍨⍤1~∧⌿∧\' '=⍵}
    trimr←⌽triml⌽
    trim←triml trimr
    
    ⍝ draw points and lines
    line←{
        0≠⎕NC'⍺':∇ ⍺ scale ⍵
        ⊃,¨/2{
            h w←-/¨y x←↓⍉↑⍺ ⍵ ⋄ h>⍥|w:⌽⍺∇⍥⌽⍵ ⋄ w>0:⍵∇⍺
            rx←⊃{0>⍵-⍺:0↑⍵ ⋄ ⍺,⍺+(1-⎕IO)+⍳⍵-⍺}/x
            ((⊃y)+⌊0.5+(⎕IO-⍨⍳≢rx)×h÷w)rx
        }/↓⍉↑⍵
    }
    _draw_←{
        ⍺←1 nice ⍺⍺ ⋄ yx←((3↑¨⍺),¨⍴⍵)scale ⍺⍺
        _point←{⊖⍺⍺@(⊃,¨/⍺)⊖⍵} ⋄ point←(⊃⍵⍵)_point
        1=≢⍵⍵:yx point ⍵
        yx point⍣(≠/⍵⍵)⊢(line yx)((2⊃⍵⍵)_point)⍵
    }

    ⍝ filling
    _fill←{
        op←⍺⍺ ⋄ (0=⎕NC'⍺')∧3≠⎕NC'⍺⍺':(2↑1 nice,⍵)∇⍵
        ⍺←' ' ⋄ fill←⍺∘{⍉⍺∘{u←⍺,∪⍵ ⋄ u[⌈\u⍳⍵]}⍤1⍉⍵}
        (,'↓')≡⎕CR'op':fill ⍵
        (,'↑')≡⎕CR'op':⊖fill⊖⍵
        (,'⊢')≡⎕CR'op':⍉fill⍉⍵
        (,'⊣')≡⎕CR'op':⍉⊖fill⊖⍉⍵
        ⍺⍺[(⍴⍵)⍴((3↑¯2↑⍺),≢⍺⍺)scale,⍵]
    }
    hlevel←hlevels _fill
    vlevel←vlevels _fill
    shade←shades _fill
    shade1←shades1 _fill
    
    ⍝ horizontal and vertical bars
    _bar_←{
        ⍺←2↑1 nice 0,⍵ ⋄ 1<≢⍵:⍺∘∇¨⍵
        n←≢⍵⍵ ⋄ f p←0 n⊤⎕IO-⍨⍵ scale⍨(¯2↑⍺),0,n×⍺⍺
        ⍉⍪(⊃⍵⍵)(⊢,⍴⍨∘(0⌈⍺⍺-≢))(f/⊃⌽⍵⍵),⍨⍣(p>0)⍨(,⍵⍵)[⎕IO+p]
    }    
    _hbar←_bar_ hlevels
    _vbar←{⍺←⊢ ⋄ ⊃⍣(⍬≡⍴⍵)⊖∘⍉¨⍺(⍺⍺ _bar_ vlevels)⍵}
    hbar←{
        ⍺←⊢ ⋄ s←⍺{0=⎕NC'⍺':scale ⍵ ⋄ 4>≢⍺:⍺ scale ⍵ ⋄ ⍺}⍵
        ⊃⍪/((2↑s)+¯1 1×((-/⌽2↑s)÷1-⍨⊃⌽s)÷2)((⊃⌽s)_hbar)⍵
    }
    vbar←{
        ⍺←⊢ ⋄ s←⍺{0=⎕NC'⍺':scale ⍵ ⋄ 4>≢⍺:⍺ scale ⍵ ⋄ ⍺}⍵
        ⊃,/((2↑s)+¯1 1×((-/⌽2↑s)÷1-⍨⊃⌽s)÷2)((⊃⌽s)_vbar)⍵
    }
    barplot←{
        ⍺←⊢ ⋄ n←⊃⌽⊃s←⍺{0=⎕NC'⍺':scale ⍵ ⋄ 2>≡⍺:⍺ scale ⍵ ⋄ ⍺}⍵
        y x←s scale ⍵ ⋄ y←(2↑⊃s){⍵ scale⍨⍺,1,(n-1)×≢vlevels}⊃⍵
        l h←(⌊/,⌈/)y←((¯1↑⊃),⍨(1↑⊃),∘{⊃⌈⌿⍤⊢⌸/⌽⍵}¯1↓¨1↓¨⊢)line y x
        (⊃⌽⊃⌽s)↑[⎕IO+1]⊃,/(l h+¯1 1×((h-l)÷n-1)÷2)(n _vbar)y
    }
    
    ⍝ ticks
    _ticks_←{t n←2↓⍵ ⋄ ⍺←⌊(n-1)÷t ⋄ ⍺⍺,⍨(n-1)⍴⍺⍺,(⍺-1)⍴⍵⍵}
    ⍝ rulers
    hruler←{⍺←⊢ ⋄ ⍺('|'_ticks_'''')⍵}
    vruler←{⍺←⊢ ⋄ ⍪⍺('—'_ticks_'-')⍵}
    ⍝ axes
    hin←{⍺←⊢ ⋄ ⍺('┴'_ticks_'─')⍵}
    vin←{⍺←⊢ ⋄ ⍪⍺('├'_ticks_'│')⍵}
    hout←{⍺←⊢ ⋄ ⍺('┬'_ticks_'─')⍵}
    vout←{⍺←⊢ ⋄ ⍪⍺('┤'_ticks_'│')⍵}
    ⍝ labels
    fmt←{
        ⍺←10 ⋄ 1>≢⍵:⍺∘∇¨⍵
        ↓⍵⎕FMT⍨⍺{1≤w←|-/2↑⍵:'I',⍕⍺ ⋄ 'F',(⍕⍺),'.',⍕-⌊10⍟w}⍵
    }
    hlabel←{
        l h t n←⍵ ⋄ ⍺←⌊(n-1)÷t
        triml⍉⍪⊃,/⍺ fmt l+(t÷⍨h-l)×⎕IO-⍨⍳1+t
    }
    vlabel←{
        (l h t n)(_ _ tx nx)←⍵ ⋄ ⍺←(⌊(n-1)÷t)(⌊(nx-1)÷tx)
        l←(⊃⌽⍺)fmt l+(t÷⍨h-l)×⎕IO-⍨⍳1+t
        triml⊖⊃⍪/⍉∘⍪¨(-1-⍨⊃⍺)↓,(⊃⍺)↑[1+⎕IO]⍪l
    }
    ⍝ 2D
    in←{stack[(vin⊃⍺ ⋄ ⍵) ⋄ ('└' ⋄ hin⊃⌽⍺)]}
    out←{stack[(vout⊃⍺ ⋄ ⍵) ⋄ ('└' ⋄ hout⊃⌽⍺)]}
    ruler←{stack[(vruler⊃⍺ ⋄ ⍵) ⋄ (' ' ⋄ hruler⊃⌽⍺)]}
    label←{stack[(vlabel ⍺ ⋄ ⍵) ⋄ (' ' ⋄ hlabel⊃⌽⍺)]}

    ⍝ plotting
    _bg←{
        ⍺←⊢ ⋄ 3>≡⍵:∇⊂⍺ scale ⍵
        ⍺⍺≡0+⎕IO:white ⍵
        ⍺⍺≡1+⎕IO:dotgrid ⍵
        ⍺⍺≡2+⎕IO:grid ⍵
        s←⊃⍵
        ⍺⍺≡3+⎕IO:s label dots2 ⍵
        ⍺⍺≡4+⎕IO:s label dotgrid ⍵
        ⍺⍺≡5+⎕IO:s label grid ⍵
        ⍺⍺≡6+⎕IO:s label s ruler dots2 ⍵
        ⍺⍺≡7+⎕IO:s label s out white ⍵
        ⍺⍺≡8+⎕IO:s label s in grid ⍵
    }
    _plot_←{
        ⍺←⊢ ⋄ 2|⎕DR ⍺⍺:⍺((⍺⍺⊃marker)∇∇ ⍵⍵)⍵
        s←⍺ scale(⊃,¨/)⍣(2<≡⍵)⊢⍵
        d←⍵ ⋄ nd←1⊣⍣(2=≡d)≢d ⋄ m←nd⍴⊆⍺⍺
        D←{
            2<≡d:⊃{s((⍺⊃d)_draw_(⍺⊃m))⍵}/(⍳nd),⊂⍵
            s(d _draw_(⊃m))⍵
        }
        p←↓_fill⍣(∧/(⊃∘⌽¨m)∊shades)D white⊂s
        p _ontop ⍵⍵ _bg⊂s
    }
    _plot←{⍺←⊢ ⋄ s←⍺ scale ⍵ ⋄ (s barplot ⍵)_ontop ⍺⍺ _bg⊂s}
    _series←{
        ⍺←12 ⋄ s←⍺{0=⎕NC'⍺':scale ⍵ ⋄ 4>≢⍺:⍺ scale ⍵ ⋄ ⍺}⍵
        (s vbar ⍵)_ontop(-⌊(⍺⍺-⎕IO)÷3)↓⍺⍺ _bg⊂s((⎕IO+0(≢⍵)),1,≢⍵)
    }
    _hbars←{
        ⍺←⊢ ⋄ names x←⍵ ⋄ s←⍺ scale x
        bg←(|⍺⍺){
            ⍺≡0+⎕IO:¯1↓white ⍵
            ⍺≡1+⎕IO:¯1↓dotgrid ⍵
            ⍺≡2+⎕IO:¯1↓grid ⍵
            ⍺≡3+⎕IO:¯1↓dots2 ⍵
            ⍺≡4+⎕IO:(hlabel s)⍪⍨¯1↓dotgrid ⍵
            ⍺≡5+⎕IO:(hlabel s)⍪⍨¯1↓grid ⍵
            ⍺≡6+⎕IO:((hruler⍪hlabel)s)⍪⍨¯1↓dots2 ⍵
            ⍺≡7+⎕IO:((hout⍪hlabel)s)⍪⍨¯1↓white ⍵
            ⍺≡8+⎕IO:((hin⍪hlabel)s)⍪⍨¯1↓grid ⍵
        }⊂(1,(⊢,(⌈3÷⍨⊢),1+⊢)≢x)s
        p←(s hbar x)_at_(2⍴⎕IO)⊢bg
        ⍺⍺≡0:p
        n←⍕⍪(-⍤≢¨⌊⌿⍛(↑¨)⊢)names
        ⍺⍺<0:stack n p ⋄ stack n p(⍕⍪x)
    }
    _histogram←{
        ⍺←50 ⋄ 2>≢⍺:(⍺,⍨⌊⍺×7÷16)∇ ⍵
        y x←⍵ ⋄ w←2×1⌈⌊(⊃⌽⍺)÷2×≢y
        sx←(⌊/,⌈/,∘(⊢,1+w×⊢)1-⍨≢)x
        sy←(⊃⍺)scale 0,y
        p←⊃,/' ',⍨¨(w-1)/¨⍪¨↓⍉sy vbar y
        p _ontop' '@(''''∘=)⍺⍺ _bg⊂sy sx
    }
    _map←{p←⍺⍺ ⍵ ⋄ 0=⎕NC'⍺':p ⋄ ⍺←(⌊/,⌈/,1,≢)∘⍳¨⍴⍵ ⋄ ⍺(⊣label ruler)p}
    plot←{
        ⍺←⊢ ⋄ s←6+⎕IO
        (1=≡⍵)∧2>≢⍴⍵:⍺(s _series)⍵
        0=80|⎕DR⊃⊃⍵:⍺(s _hbars)⍵
        2<≡⍵:⍺(((≢⍵)⍴2⍴¨⌽2↓shades1)_plot_ s)⍵
        2=≢⍴⍵:⍺ shade1 _map ⍵
        ≠/≢¨⍵:⍺(s _histogram)⍵
        ⍺(s _plot)⍵
    }
    
    ⍝ tables
    hline←{⍺←1 ⋄ ⍺<0:⊖(|⍺)∇⊖⍵ ⋄ ⍺(↑⍪'─'⍪↓)⍵}
    undup←{⍺←' ' ⋄ ⍉⍺@(0,2≡/⊢)⍤1⍉⍵}
    nan←{⍺←dot ⋄ (⊂⍺)@(≡∘⍬¨)⍵}
    table←{t←⍪¨⍵ ⋄ 0=⎕NC'⍺':⍕t ⋄ hline⍕(⊂¨⍺)⍪¨t}
    _maxlines_←{
        ⍺←⊢ ⋄ d←⍺ ⍺⍺⊢(⍵⍵⌊n←≢⊃⍵)∘↑¨⍵
        (⍵⍵+2){↑(⊂'…',⍕n),⍨⍣(⍺<n+2)↓(n⌊⍺)↑⍵}d
    }
    
    ⍝ block drawing
    tesselate←{⊂[2×⍳k](,s,⍪r)⍴⍵↑⍨r×s←⌈R÷r←1+(-k←≢R←⍴⍵)↑⍺-1} ⍝ aplcart
    _block_←{
        0=⎕NC'⍺':⍺⍺[⎕IO+2⊥∘,¨2 1 tesselate ⍵] ⋄ 0>⍺:(|⍺)∇⌽⍵ ⋄ 1=⍺:∇ ⍵
        u d←60{(⊢÷∨/)(⌊0.5+⍵×⍺)⍺}⍺ ⋄ g←u{⍺⌿⍺/⍵}⍣(u>1)⊢⍵ ⋄ 1=d:∇ g
        ((⌊/,⌈/)⍤∊⍵⍵ _fill⊢)+/∘∊¨(|(d×2)d)tesselate g
    }
    block←blocks _block_ shades
    shaded←shades _block_ shades
    zoom←{
        ⍺←2 ⋄ 1=⍺:⍵ ⋄ 0>⍺:(|⍺)∇⌽⍵
        u d←60{(⊢÷∨/)(⌊0.5+⍵×⍺)⍺}⍺ ⋄ g←u{⍺⌿⍺/⍵}⍣(u>1)⊢⍵ ⋄ 1=d:g
        (⊢>2÷⍨⌈/∘∊)+/∘∊¨(|d d)tesselate g
    }
    dotted←{
        ⍺←1 ⋄ v←¯2(↑,⍨∘,∘⍉2 3⍴↓)2*⎕IO-⍨⍳8
        braille[⎕IO+(v+.×,)¨4 2 tesselate ⍺ zoom ⍵]
    }


    ⍝ https://www.dyalog.com/the-dyalog-duck.htm
    duck←24 40⍴⊃,/≢⍛⍴∘0 1⍛(/¨⍨)⊃,/(
        66 5 34 8 32 9 30 11 29 12 28 15 24 15 25 8 32 6 34 6 26 14 23 18 11
        1 8 21 10 4 1 26 9 32 9 31 10 31 10 30 11 29 12 27 15 25 20 18 49
    )

:EndNamespace
```

## Scaling of coordinates

```apl
⍝ data
2⍕↑yx
```

### Nice

```apl
⍝ with a vector, returns low boundary, high boundary and number of intervals with nice (round numbers) limits
nice x
```

```apl
⍝ with nested vectors, returns a nice scale for each vector
nice yx
```

```apl
⍝ dyadically, specify approximate number of intervals
10 nice x ⋄ 20 30 nice yx
```

### Scale

```apl
⍝ monadically, scale returns an scale with an additional element indicating the total length
scale x
scale yx
```

```apl
⍝ a numeric left argument can specify an approximate length, or width for nested vectors,
⍝ or both width and height, if the right argument is a pair of vectors
30 scale yx ⋄ 30 30 scale yx
```

```apl
⍝ giving the result of scale as left argument will return the indices corresponding to the data at right
scale⍛scale x ⋄ scale⍛scale yx
```

```apl
⍝ original coordinates and indices for two different scales
⍉∘↑¨yx(scale⍛scale yx)(50∘scale⍛scale yx)
```

## Rectangles

```apl
⍝ create rectangles with predefined backgrounds of the given shape
2 4⍴{(white ⍵)(light ⍵)(dark ⍵)(black ⍵)(dots ⍵)(dots2 ⍵)(hdots ⍵)(lines ⍵)}7 20
```

```apl
⍝ you can also directly pass data and optional scale parameters, or enclosed scales,
⍝ to create a rectangle of a sensible size to fit the data
⍕(black yx)(30 black yx)(black⊂20 scale yx)
```

```apl
⍝ grid, dotgrid and dgrid create grids with the size specified as left argument (2 2 by default)
2 3⍴{(grid ⍵)(dotgrid ⍵)(dgrid ⍵)(3 5 grid ⍵)(3 5 dotgrid ⍵)(3 5 dgrid ⍵)}7 20
```

```apl
⍝ or using scales
⍕{(grid ⍵)(30 grid ⍵)(grid⊂20 scale ⍵)}yx
```

```apl
⍝ more generally, the _grid_ operator can be used to create other grid types
(30('+'_grid_'-|')yx)(30('∘'_grid_ dot)yx)
```

```apl
⍝ and the _rect operator repeats the array given as left operand, reshaping it first when called dyadically
2 4{(⍺(⎕A _rect)⍵)('⎕'_rect ⍵)((box 0 1)_rect ⍵)}7 20
```

### Boxes

```apl
⍝ dyadically, it will first reshape the array given at right with the shape given at left
3 5 box ⎕A
```

## Combine layers

```apl
⍝ data
h m b
```

```apl
⍝ the _ontop operator overlays its left operand on top of its right argument, aligned at the top right
⍝ the _at_ operator allows more specific positioning, taking coordinates as right argument
⍝ both operators can take an optional left argument indicating the "alpha channel" (' ' by default)
(h _ontop b)(h _at_ 3 5⊢b)(h _at_ 7 9⊢b)(h _at_ 0 ¯1⊢b)(m _ontop⊢b)('0'(m _ontop)b)('0'(m _at_ 2 2)b)
```

```apl
⍝ stack composes the elements in the given array aligning smaller ones to the top right corner,
⍝ the optional left argument can be used to change this alignment
(stack h b)(stack⍪h b),{(stack ⍵)(¯1 1 stack ⍵)(1 1 stack ⍵)}[('a' ⋄ h) ⋄ (m ⋄ b)]
```

## Tables

```apl
⍝ table creates a table with the data given as right argument and the column names given as left argument
columns table id name(12 _hbar age)col dob
```

```apl
⍝ column names are optional
table id name(12 _hbar age)col dob
```

```apl
⍝ the function undup display as blanks consecutive duplicated lines (useful to list categories, for example)
⍝ nan will substitute missing data with a dot
⍝ the _maxlines_ operator will limit the number of lines and display the total if larger
⍝ hline adds horizontal lines after the given line (the first one by default, negative to count from end)
ms1 ms2 ms3←(months,↓vlevel sales)(months,↓hlevel sales)(months,↓(shades1,'⎕')_fill sales)
captions←'n' 'x' 'xh' 'y' 'yv' 'letter' 'monthly1' 'monthly2' 'monthly3'
¯1 hline captions(table _maxlines_ 25)(undup nan id2)x1(hlevel x1)y1(hlevel y1)abc ms1 ms2 ms3
```

```apl
⍝ a left argument to undup or nan can be used to use other replacement
⍝ notice that _maxlines_ only runs the left function with the selected lines
captions({⎕←≢⊃⍵ ⋄ ⍺ table ⍵}_maxlines_ 10)('"'undup'NaN'nan id2)x1(hlevel x1)y1(hlevel y1)abc ms1 ms2 ms3
```
